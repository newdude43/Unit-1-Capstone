The data set represented is a summary of Kickstarter.com projects from May 03, 2009 through March 03, 2018.

### Analysis that highlights your experimental hypothesis

I hypothesize that a new, paid tier system that gives users better website visibility will result in an increased project success rate.  The paid tier system would place participating projects on the front page of Kickstarter.com 

### A rollout plan showing how you would implement and rollout the experiment

Kickstarter currently determines featured projects by several factors.  Here is a link on how to get featured on Kickstarter: https://www.kickstarter.com/blog/how-to-get-featured-on-kickstarter

With our experiment, we would need to set parameters to make sure results are fair and unbiased. By determining a main category to focus in on, we could pick a random sample of 30-50 projects and feature them on the front page of Kickstarter.com.  


### An evaluation plan showing what constitutes success in this experiment

-**Gather** 30-50 random projects

-Feature those projects on the front page of Kickstarter.com during each respective project's campaign period

-Run T test for independence to see how confident we can be with the test results

-**Success** is defined by:

1st - meeting project goal

2nd - money received divided by total money goal

-We would expect the success rate to be higher for the projects featured on the front page.  If this is not the case, then we will need to check out the individual projects tested to see if there were any bad apples or outlying factors that would have factored into our findings.

-After comparing success rates, we will want to check out money brought in divided by total fundraising money goal.  For example, bringing in 100 USD for a 100 USD goal would equal 1.0.

## Average Backers Per Main Category

```python
with plt.style.context('fivethirtyeight'):
    likes_plot = df.groupby('main_category')['backers'].mean().plot(kind = 'bar', figsize = [10, 6])
    likes_plot.set_xlabel('Main Category')
    likes_plot.set_ylabel('Average No. of Backers')
    likes_plot.set_title('Average Backers Per Main Category')
    plt.xticks(rotation=90)
    plt.show()
```

## Let us focus in on the Games category
**Games** has a large number of backers, pledges, and total successes; however, it has a pretty low success rate (36%).  If we can isolate the testing and focus in on only Games, that will isolate variables and help us determine if our test is successful.

## The Test
In the Games category, we will perform the following:

-**Gather** 30-50 random projects

-Feature those projects on the front page of Kickstarter.com during the respective project's campaign period

-Currently Games has a success rate of 36% or 0.36. 

-After test period is complete, compare success rate of sample group vs. Games' current 36% and a separate comparison vs. Games projects not included on the front page (part of the 36% may be featured on the front page already via Kickstarter's normal choosing parameters)

-Run T test for independence to see how confident we can be with the test results

1st - meeting project goal

2nd - money received divided by total money goal
